```python
df1 = pd.read_csv("2022_Problem_C_DATA/BCHAIN-MKPRU.csv")
# print(df1['Value'])
y = np.random.standard_normal(size=1550)

for i in range(0,1550):
        y[i] = df1['Value'][i]

plt.figure(figsize=(12,6))
plt.plot(y)
plt.show()
```

```python
from arch.unitroot import ADF
# ADF 平稳性分析ADF方法
adf = ADF(y)
# print(adf.pvalue)
print(adf.summary().as_text())
# 检验统计量 -1.002
# P 值 0.753
# 滞后 10
# 趋势：持续
# 临界值： -3.43 （1%）， -2.86 （5%）， -2.57 （10%）
# 原假设：进程包含单位根。
# 替代假设：该过程是弱静止的。
```
