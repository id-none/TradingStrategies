```python
# Python
import pandas as pd
import numpy as np
from fbprophet import Prophet
```

```python
df = pd.read_csv('2022_Problem_C_DATA/LBMA-GOLD_re.csv',encoding='utf-8')
# df['y'] = np.log(df['y'])
df.head()
m = Prophet()
# m = Prophet(growth='logistic')
# df['cap'] = 6
m.fit(df)
# Python
# future = m.make_future_dataframe(periods=365)
future = m.make_future_dataframe(periods=365) 
# Python
# future['cap'] = 5
forecast = m.predict(future)
forecast.to_excel('2022_Problem_C_DATA/LBMA-GOLD_pre.xlsx')
m.plot(forecast)
m.plot_components(forecast).show()
```
