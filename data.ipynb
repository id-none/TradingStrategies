```python
def return_time(text):
    ls = text.split('/')
    s = ""
    s += "20"+ls[2] + "."
    s += ls[0] + "."
    s += ls[1]
    return s
```

```python
print(return_time("9/11/16"))
```

```python
f = open('2022_Problem_C_DATA/LBMA-GOLD_0.txt','r',encoding="utf-8")
f1 = open('2022_Problem_C_DATA/LBMA-GOLD_1.txt','w',encoding="utf-8")
text = f.readlines()
text = [line.strip("\n") for line in text]
print(len(text))
n = 0
for i in text:
    f1.write(return_time(i)+"\n")
    n = n+1
print(n)
    
```
