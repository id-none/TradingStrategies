```python
shar = []
for i in range(30,len(df),30):
    rsyls = []
    for j in range(30):
        rsyls.append(df['syl'][i-j])

    rsyls_average = np.mean(rsyls)
    rsyls_biaozhucha = np.std(rsyls)
    # print(str(rsyls_average) + "     " +str(rsyls_biaozhucha) )
    # print(rsyls_biaozhucha)
    shar.append((rsyls_average-0.03)/rsyls_biaozhucha)

for i in shar:
    print(i)
```

```python
import pandas as pd
import numpy as np

df1 = pd.read_excel("LBMA-GOLD_re_alltime.xlsx")
df2 = pd.read_excel("LBMA-GOLD_re_alltime.xlsx",sheet_name="Sheet1")
df1.fillna(method='ffill',inplace=True)
df1.to_csv("ffill.csv")
print(df1.head())
df2.head()
```

```python
import numpy as np
fx = []
for i in df2['风险']:
    fx.append(i)

# print(fx)
p = []
for i in range(30,len(df1),30):
    pi =[]
    for j in range(0,30):
        pi.append(float(df1['USD (PM)'][i-j]))
    p.append(np.mean(pi))
print(p)

k = []
for i in range(len(p)):
    k.append(1/p[i])
print(k)
```

```python
g=[]
g.append(400/df1['USD (PM)'][0])
for i in range(1,len(k)):
    if df1['USD (PM)'][(i-1)*30] > df1['USD (PM)'][(i-1)*30]:
        g.append(0.95* k[i-1] * (fx[i]-fx[i-1]) + g[i-1])
    else:
        g.append(1.05* k[i-1] * (fx[i]-fx[i-1]) + g[i-1])

for i in range(len(g)):
    print(g[i])

flag = True
for i in range(len(g)-1):
    if  np.abs(g[i+1] - g[i]) < 0.01:
        flag = False
print(flag)
```
