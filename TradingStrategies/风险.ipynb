```python
bitebi_average = []
for i in range(90,len(df),90):
    syl = []
    for j in range(90):
        syl.append(df['syl'][i - j])
    bitebi_average.append(np.mean(syl))
print(bitebi_average)
```

```python
df = pd.read_csv("ffill.csv")
df.head()
gold_average = []
for i in range(90,len(df),90):
    syl = []
    for j in range(90):
        n = 90
        if df['syl'][i - j] == 0:
            n = n-1
        syl.append(df['syl'][i - j])
    gold_average.append(np.sum(syl)/n)
print(gold_average)
```

```python
g = []
g.append([0,1000])
bili = []
for i in range(20):
    if bitebi_average[i] > 0 and bitebi_average[i]/gold_average[i] < 0:
        bili.append(6/4)
    elif bitebi_average[i] < 0 and bitebi_average[i]/gold_average[i] < 0:
        bili.append(4/6)
    elif bitebi_average[i] < 0 and gold_average[i] < 0:
        bili.append(gold_average[i]/bitebi_average[i])
    else:
        bili.append(bitebi_average[i]/gold_average[i])
print(bili)

```
